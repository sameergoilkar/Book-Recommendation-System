# Merge Ratings on Books

```python
rb_new=ratings.merge(books,on='ISBN')
```

```python
rb_new.head()
```

```python
rb_n_ratings = rb_new.groupby(['Book-Title']).count()['Book-Rating'].reset_index()
```

```python
rb_n_ratings.rename(columns={'Book-Rating' : 'n_ratings'},inplace=True)
```

```python
rb_n_ratings
```

```python
rb_avg_ratings = rb_new.groupby(['Book-Title']).mean()['Book-Rating'].reset_index()
rb_avg_ratings.rename(columns={'Book-Rating' : 'avg_ratings'},inplace=True)
```

```python
n_avg_ratings= rb_n_ratings.merge(rb_avg_ratings,on='Book-Title')
```

```python
n_avg_ratings
```

```python
popular_df = n_avg_ratings[n_avg_ratings['n_ratings']>250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','n_ratings','avg_ratings']]
```

```python
popular_df
```

# Collaborative Filtering Based Recommender System

```python
x = rb_new.groupby('User-ID').count()['Book-Rating']>200
max_users = x[x].index
```

```python
filtered_ratings = rb_new[rb_new['User-ID'].isin(max_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

# Similarity Score

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
